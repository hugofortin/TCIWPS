```python
def analyze_hodge_wps(alpha_b, d, weights):
    m = [d // wi for wi in weights]     # per-coordinate orders
    N = 36
    K_group = [k for k in range(1, N) if gcd(k, N) == 1]

    n = len(weights) - 2               # hypersurface dimension

    print(f"{'k':<4} | {'a_prime':<22} | {'s':<6} | {'p':<2} | Hodge")
    print("-"*60)

    for k in K_group:
        a_prime = []
        s = QQ(0)

        for bi, mi in zip(alpha_b, m):
            ai = bi + 1                            # <-- the crucial fix
            r = (k * ai) % mi
            if r == 0:
                r = mi                             # keep 1..mi
            a_prime.append(r)
            s += QQ(r) / QQ(mi)

        p = s - 1
        if p in ZZ and 0 <= p <= n:
            p = int(p)
            q = n - p
            hodge = f"H{p},{q}"
        else:
            hodge = "not residue type"

        print(f"{k:<4} | {str(a_prime):<22} | {str(s):<6} | {str(p):<2} | {hodge}")

```

```python
from itertools import product

try:
    import scipy.sparse as sp
    from scipy.sparse import save_npz
except ImportError:
    sp = None
    save_npz = None


# ------------------------------------------------------------
# 1) Build the basis in EXACTLY the same order as the period matrix
#
#    all_van_beta order:
#      [0,b1,b2,b3,b4,b5]
#    with
#      b1,b2 in range(35), b3 in range(3), b4,b5 in range(2)
# ------------------------------------------------------------

def weighted_degree36_betas_in_user_order():
    """
    Returns
    -------
    all_van_beta : list of 6-tuples
        In exactly the same order as the period matrix rows.

    betas_affine : list of 5-tuples
        The corresponding affine Brieskorn-Pham labels (b1,b2,b3,b4,b5)
        in the same order.
    """
    all_van_beta = []
    betas_affine = []

    for b1 in range(35):
        for b2 in range(35):
            for b3 in range(3):
                for b4 in range(2):
                    for b5 in range(2):
                        all_van_beta.append((0, b1, b2, b3, b4, b5))
                        betas_affine.append((b1, b2, b3, b4, b5))

    return all_van_beta, betas_affine


# ------------------------------------------------------------
# 2) Movasati pairing for the affine Brieskorn-Pham fiber
#
#    x1^36 + x2^36 + x3^4 + x4^3 + x5^3 = 1
#
#    n = 4, so:
#      <δβ,δβ> = 2
#      <δβ,δβ'> = (-1)^(sum(beta'_i - beta_i))
#        if each difference is 0 or 1
#      symmetric
#      0 otherwise
# ------------------------------------------------------------

def movasati_affine_pair_n4(beta, beta_prime):
    beta = tuple(beta)
    beta_prime = tuple(beta_prime)

    if len(beta) != 5 or len(beta_prime) != 5:
        raise ValueError("beta and beta_prime must both be length-5 tuples")

    if beta == beta_prime:
        return 2

    diff = [bp - b for b, bp in zip(beta, beta_prime)]
    if all(d in (0, 1) for d in diff):
        return -1 if (sum(diff) % 2) else 1

    diff = [b - bp for b, bp in zip(beta, beta_prime)]
    if all(d in (0, 1) for d in diff):
        return -1 if (sum(diff) % 2) else 1

    return 0


# ------------------------------------------------------------
# 3) Build Psi in THAT exact order
# ------------------------------------------------------------

def build_affine_intersection_matrix_user_order(sparse=True):
    """
    Build the affine 14700 x 14700 intersection matrix Psi in exactly the
    same order as the user's period matrix:
        [0,b1,b2,b3,b4,b5].

    Returns
    -------
    all_van_beta : list of 6-tuples
    betas_affine : list of 5-tuples
    Psi          : sparse matrix or dict
    """
    if sparse and sp is None:
        raise ImportError("scipy is required for sparse=True")

    exponents = (36, 36, 4, 3, 3)
    all_van_beta, betas_affine = weighted_degree36_betas_in_user_order()

    N = len(betas_affine)   # should be 14700
    idx = {beta: i for i, beta in enumerate(betas_affine)}

    rows = []
    cols = []
    data = []

    d = len(exponents)
    eps_list = [eps for eps in product((0, 1), repeat=d) if any(eps)]

    # diagonal
    for i in range(N):
        rows.append(i)
        cols.append(i)
        data.append(2)

    # neighbors
    for beta in betas_affine:
        i = idx[beta]
        for eps in eps_list:
            beta2 = tuple(beta[k] + eps[k] for k in range(d))
            if any(beta2[k] > exponents[k] - 2 for k in range(d)):
                continue

            j = idx[beta2]
            if i < j:
                val = -1 if (sum(eps) % 2) else 1
                rows.append(i)
                cols.append(j)
                data.append(val)

                rows.append(j)
                cols.append(i)
                data.append(val)

    if sparse:
        Psi = sp.coo_matrix((data, (rows, cols)), shape=(N, N)).tocsr()
    else:
        Psi = {(i, j): v for i, j, v in zip(rows, cols, data)}

    return all_van_beta, betas_affine, Psi


# ------------------------------------------------------------
# 4) Optional save helpers
# ------------------------------------------------------------

def save_labels_txt(labels, filename):
    with open(filename, "w") as f:
        f.write("[\n")
        for i, row in enumerate(labels):
            if i > 0:
                f.write(",\n")
            f.write(str(list(row)))
        f.write("\n]\n")


# ------------------------------------------------------------
# 5) Build Psi 
# ------------------------------------------------------------

if __name__ == "__main__":
    all_van_beta, betas_affine, Psi = build_affine_intersection_matrix_user_order(sparse=True)

    print("number of vanishing cycles =", len(all_van_beta))   # 14700
    print("first label =", all_van_beta[0])                    # (0,0,0,0,0,0)
    print("second label =", all_van_beta[1])                   # (0,0,0,0,0,1) in this ordering
    print("last label =", all_van_beta[-1])                    # (0,34,34,2,1,1)

    print("first affine beta =", betas_affine[0])
    print("second affine beta =", betas_affine[1])

    print("Psi shape =", Psi.shape)
    print("Psi nnz   =", Psi.nnz)

    # single-entry check
    i, j = 0, 1
    print("pair(first, second) =", movasati_affine_pair_n4(betas_affine[i], betas_affine[j]))

    # save matrix + ordering
    if save_npz is not None:
        save_npz("Psi_affine_14700", Psi, compressed=True)
        print("saved Psi to Psi_affine_14700.npz")

    save_labels_txt(all_van_beta, "all_van_beta_user_order.txt")
    save_labels_txt(betas_affine, "betas_affine_user_order.txt")
    print("saved row order to all_van_beta_user_order.txt")
    print("saved affine labels to betas_affine_user_order.txt")
```

```python
# SageMath
#
# h22 lattice pipeline over Q(zeta_36), with cycles-at-infinity quotient
#
# New feature:
#   after building the affine lattice B_aff and affine Gram matrix G_aff,
#   quotient by the saturated integral radical of G_aff.
#
# This implements the "remove cycles at infinity inside the candidate lattice"
# step.
#
# Files:
#   matrixdegree36homologyh22even.txt
#   Psi.npz
#
# Outputs:
#   B_affine.txt, G_affine.txt, G_affine.sobj
#   B_projective.txt, G_projective.txt, G_projective.sobj
#
# If the only reason G_aff is not positive definite is the affine radical,
# then G_projective is the corrected form to inspect.

from sage.all import *
import ast
import os

try:
    import scipy.sparse as sp
    from scipy.sparse import load_npz
except ImportError:
    sp = None
    load_npz = None


# ------------------------------------------------------------
# 1) Stream rows from one huge list-of-lists txt file
# ------------------------------------------------------------

def row_strings_from_big_list_file(filename, chunk_size=1 << 20):
    with open(filename, "r") as f:
        depth = 0
        buf = []
        in_row = False

        while True:
            chunk = f.read(chunk_size)
            if not chunk:
                break

            for ch in chunk:
                if ch == '[':
                    depth += 1
                    if depth == 2:
                        in_row = True
                        buf = ['[']
                    elif in_row:
                        buf.append(ch)

                elif ch == ']':
                    if in_row:
                        buf.append(']')
                    depth -= 1
                    if in_row and depth == 1:
                        in_row = False
                        yield ''.join(buf)

                else:
                    if in_row:
                        buf.append(ch)


# ------------------------------------------------------------
# 2) Load period matrix P over K = Q(zeta_36)
# ------------------------------------------------------------

def load_period_matrix_over_K_from_txt(filename, K, z, chunk_size=1 << 20, verbose=True):
    rows = []
    ncols = None
    processed = 0
    eval_locals = {'z': z, 'I': I, 'sqrt': sqrt, 'K': K}

    for row_str in row_strings_from_big_list_file(filename, chunk_size=chunk_size):
        row = sage_eval(row_str, locals=eval_locals, preparse=True)
        row = [K(a) for a in row]

        if ncols is None:
            ncols = len(row)
        elif len(row) != ncols:
            raise ValueError("Inconsistent row lengths in {}".format(filename))

        rows.append(row)
        processed += 1

        if verbose and processed % 1000 == 0:
            print("loaded period rows:", processed)

    return matrix(K, rows) if rows else matrix(K, 0, 0)


# ------------------------------------------------------------
# 3) Load sparse Psi
# ------------------------------------------------------------

def load_Psi_npz_checked(filename, expected_n, verbose=True):
    if load_npz is None:
        raise ImportError("scipy is required to read .npz sparse matrices")

    A = load_npz(filename).tocsr()

    if verbose:
        print("Psi scipy shape:", A.shape, "| nnz:", A.nnz)

    if A.shape != (expected_n, expected_n):
        raise ValueError(
            "Psi has shape {}, but period matrix has {} rows.".format(A.shape, expected_n)
        )

    return A


def scipy_csr_to_sage_sparse_int(A):
    C = A.tocoo()
    entries = {
        (int(i), int(j)): ZZ(v)
        for i, j, v in zip(C.row, C.col, C.data)
        if v != 0
    }
    return matrix(ZZ, int(A.shape[0]), int(A.shape[1]), entries, sparse=True)


# ------------------------------------------------------------
# 4) Expand P over the power basis [1,z,...,z^(d-1)]
# ------------------------------------------------------------

def expand_K_matrix_to_QQ(P):
    P = matrix(P)
    K = P.base_ring()

    if K == QQ:
        return P.change_ring(QQ)

    d = K.degree()
    mu = P.nrows()
    r = P.ncols()

    Ptilde = matrix(QQ, mu, r * d)

    for i in range(mu):
        for j in range(r):
            coeffs = list(K(P[i, j]).list())
            if len(coeffs) < d:
                coeffs += [QQ(0)] * (d - len(coeffs))
            elif len(coeffs) > d:
                coeffs = coeffs[:d]

            for ell in range(d):
                Ptilde[i, j * d + ell] = QQ(coeffs[ell])

    return Ptilde


def check_expansion(P, Ptilde, K, z, num_rows=3, num_cols=3):
    d = K.degree()
    rmax = min(num_rows, P.nrows())
    cmax = min(num_cols, P.ncols())

    for i in range(rmax):
        for j in range(cmax):
            rec = sum(K(Ptilde[i, j*d + ell]) * (z**ell) for ell in range(d))
            if rec != P[i, j]:
                print("Expansion mismatch at", (i, j))
                print("original  =", P[i, j])
                print("recovered =", rec)
                return False

    print("Expansion check passed on first", rmax, "rows and", cmax, "columns")
    return True


# ------------------------------------------------------------
# 5) Exact solve without Sage's rank precheck
# ------------------------------------------------------------

def solve_right_nonsingular_batched(PsiQ, RHS, batch_cols=8, verbose=True):
    """
    Solve PsiQ * X = RHS over QQ, assuming PsiQ is nonsingular.

    Uses the private method _solve_right_nonsingular_square(..., check_rank=False)
    to avoid Sage's expensive rank precheck.
    """
    PsiQ = matrix(QQ, PsiQ)
    RHS = matrix(QQ, RHS)

    if PsiQ.nrows() != PsiQ.ncols():
        raise ValueError("PsiQ must be square")
    if RHS.nrows() != PsiQ.nrows():
        raise ValueError("RHS has wrong number of rows")

    pieces = []
    total = RHS.ncols()

    for j0 in range(0, total, batch_cols):
        j1 = min(j0 + batch_cols, total)
        if verbose:
            print("solving RHS columns", j0, "to", j1 - 1)

        rhs_blk = RHS[:, j0:j1]
        sol_blk = PsiQ._solve_right_nonsingular_square(rhs_blk, check_rank=False)
        pieces.append(sol_blk)

    return block_matrix(1, len(pieces), pieces)


# ------------------------------------------------------------
# 6) Row-HNF helper and p-saturation helpers
# ------------------------------------------------------------

def row_hnf_basis(B, proof=False):
    B = matrix(ZZ, B)
    ncols = B.ncols()

    if B.nrows() == 0:
        return matrix(ZZ, 0, ncols)

    H = B.dense_matrix().hermite_form(include_zero_rows=True, proof=proof)
    rows = [list(r) for r in H.rows() if not r.is_zero()]

    if rows:
        return matrix(ZZ, rows)
    return matrix(ZZ, 0, ncols)


def p_saturate_rows(B, p, proof=False, verbose=True, max_rounds=20):
    B = row_hnf_basis(matrix(ZZ, B), proof=proof)

    for step in range(max_rounds):
        M = B.change_ring(GF(p))
        K = M.transpose().right_kernel_matrix()

        if K.nrows() == 0:
            if verbose:
                print(f"p={p}: already p-saturated")
            return B

        new_rows = []
        for c in K.rows():
            cZZ = vector(ZZ, [ZZ(int(a)) for a in c])
            w = cZZ * B
            if any(x % p != 0 for x in w):
                raise RuntimeError(
                    f"Internal error: lifted kernel vector did not give divisibility by {p}"
                )
            w_div = [ZZ(x // p) for x in w]
            if any(x != 0 for x in w_div):
                new_rows.append(w_div)

        if not new_rows:
            if verbose:
                print(f"p={p}: no new rows added")
            return B

        B = row_hnf_basis(B.stack(matrix(ZZ, new_rows)), proof=proof)

        if verbose:
            print(f"p={p}: round {step+1}, basis rows = {B.nrows()}")

        if B.change_ring(GF(p)).rank() == B.nrows():
            if verbose:
                print(f"p={p}: saturation complete")
            return B

    raise RuntimeError(f"p-saturation did not stabilize for p={p} within max_rounds={max_rounds}")


def clear_denominators_and_psaturate_rows(Bq, primes=None, proof=False, verbose=True):
    dens = [a.denominator() for a in Bq.list()]
    D = lcm(dens) if dens else ZZ(1)

    Bint = (D * Bq).change_ring(ZZ)
    B = row_hnf_basis(Bint, proof=proof)

    try:
        B = B.LLL()
        B = row_hnf_basis(B, proof=proof)
    except Exception:
        pass

    if primes is None:
        primes = list(ZZ(D).prime_divisors())

    if verbose:
        print("initial denominator multiplier D =", D)
        print("initial candidate saturation primes =", primes)

    for p in primes:
        B = p_saturate_rows(B, p, proof=proof, verbose=verbose)

    return B, D


def audit_small_primes(B, bound=29):
    bad = []
    for p in prime_range(2, bound + 1):
        if B.change_ring(GF(p)).rank() < B.nrows():
            bad.append(p)
    return bad


# ------------------------------------------------------------
# 7) Exact small Gram computation using scipy CSR
#    (avoids Sage dense*sparse multiplication issues)
# ------------------------------------------------------------

def right_multiply_small_basis_by_sparse_psi(B, Psi_csr):
    """
    Compute C = B * Psi_csr exactly, where
      B       is a small Sage ZZ matrix of shape (m x n)
      Psi_csr is a scipy CSR matrix of shape (n x n)

    Output is a Sage ZZ matrix of shape (m x n).
    """
    m = B.nrows()
    n = B.ncols()

    row_dicts = [dict() for _ in range(m)]

    indptr = Psi_csr.indptr
    indices = Psi_csr.indices
    data = Psi_csr.data

    for i in range(n):
        start = indptr[i]
        end = indptr[i + 1]
        if start == end:
            continue

        col_i = [B[r, i] for r in range(m)]
        if all(v == 0 for v in col_i):
            continue

        for pos in range(start, end):
            j = int(indices[pos])
            a = ZZ(int(data[pos]))
            if a == 0:
                continue
            for r in range(m):
                v = col_i[r]
                if v != 0:
                    row_dicts[r][j] = row_dicts[r].get(j, ZZ(0)) + v * a

    rows = []
    for d in row_dicts:
        row = [ZZ(0)] * n
        for j, val in d.items():
            row[j] = val
        rows.append(row)

    return matrix(ZZ, rows)


def gram_from_basis_and_sparse_psi(B, Psi_csr):
    C = right_multiply_small_basis_by_sparse_psi(B, Psi_csr)
    return C * B.transpose()


# ------------------------------------------------------------
# 8) Radical / cycles-at-infinity quotient inside the affine lattice
# ------------------------------------------------------------

def saturated_integral_radical_of_gram(G, proof=False):
    """
    Return a saturated ZZ-basis of the integral radical of the symmetric
    integer Gram matrix G.

    Output R has rows in coefficient coordinates relative to the current
    basis of the lattice.
    """
    Gq = matrix(QQ, G)
    Kq = Gq.right_kernel_matrix()   # rows = QQ-basis of radical

    if Kq.nrows() == 0:
        return matrix(ZZ, 0, G.nrows())

    dens = [a.denominator() for a in Kq.list()]
    D = lcm(dens) if dens else ZZ(1)
    Kint = (D * Kq).change_ring(ZZ)

    L = Kint.row_module(ZZ).saturation()
    R = row_hnf_basis(matrix(ZZ, L.basis_matrix()), proof=proof)
    return R


def pivot_columns_of_row_hnf(H):
    pivots = []
    for r in range(H.nrows()):
        row = H.row(r)
        for j, a in enumerate(row):
            if a != 0:
                pivots.append(j)
                break
    return pivots


def quotient_by_radical_via_unimodular_change(B_aff, G_aff, proof=False, verbose=True):
    """
    Given:
      B_aff : basis rows in affine cycle coordinates
      G_aff : affine Gram matrix on that basis

    Compute the saturated radical R of G_aff, then a unimodular basis change
    separating radical directions from quotient directions.

    Returns
    -------
    R_coeff : radical basis rows in coefficient coordinates relative to B_aff
    B_rad   : radical basis rows in affine coordinates
    B_proj  : quotient/projective basis rows in affine coordinates
    G_proj  : quotient/projective Gram matrix
    U       : unimodular change-of-basis matrix on coefficient lattice
    G_new   : transformed Gram matrix U * G_aff * U^T
    """
    n = B_aff.nrows()
    R_coeff = saturated_integral_radical_of_gram(G_aff, proof=proof)
    s = R_coeff.nrows()

    if verbose:
        print("radical rank =", s)

    if s == 0:
        return (
            R_coeff,
            matrix(ZZ, 0, B_aff.ncols()),
            B_aff,
            G_aff,
            identity_matrix(ZZ, n),
            G_aff
        )

    H = row_hnf_basis(R_coeff, proof=proof)
    pivots = pivot_columns_of_row_hnf(H)
    nonpivots = [j for j in range(n) if j not in pivots]

    # For a primitive saturated submodule, H should have pivot 1's.
    for r, j in enumerate(pivots):
        if H[r, j] != 1:
            raise RuntimeError(
                "Unexpected non-primitive radical basis: pivot entry is not 1. "
                "Check saturation step."
            )

    # Build a unimodular matrix U whose first s rows span the radical
    rows_U = [list(H.row(r)) for r in range(H.nrows())]
    for j in nonpivots:
        e = [ZZ(0)] * n
        e[j] = ZZ(1)
        rows_U.append(e)

    U = matrix(ZZ, rows_U)

    if abs(U.det()) != 1:
        raise RuntimeError("Constructed basis-change matrix is not unimodular.")

    B_new = U * B_aff
    G_new = U * G_aff * U.transpose()

    # first s rows/cols should be radical
    B_rad = B_new.matrix_from_rows(range(s))
    B_proj = B_new.matrix_from_rows(range(s, n))
    G_proj = G_new.matrix_from_rows(range(s, n)).matrix_from_columns(range(s, n))

    return R_coeff, B_rad, B_proj, G_proj, U, G_new


# ------------------------------------------------------------
# 9) Main routine
# ------------------------------------------------------------

def compute_lattice_from_h22_periods_Qzeta36_with_infinity_quotient(
    period_file="matrixdegree36homologyh22even.txt",
    psi_file="Psi.npz",
    expected_affine_rank=18,
    expected_projective_rank=None,
    saturation_primes=None,
    chunk_size=1 << 20,
    rhs_batch_cols=6,
    verbose=True,
    save_txt=True,
    save_sobj=True,
    B_aff_output="B_affine.txt",
    G_aff_output="G_affine.txt",
    G_aff_sobj="G_affine.sobj",
    B_proj_output="B_projective.txt",
    G_proj_output="G_projective.txt",
    G_proj_sobj="G_projective.sobj",
    B_rad_output="B_radical.txt",
    proof=False,
    check_basis_expansion=True,
):
    K.<z> = CyclotomicField(36)

    if verbose:
        print("loading period matrix P over Q(zeta_36) ...")
    P = load_period_matrix_over_K_from_txt(period_file, K, z, chunk_size=chunk_size, verbose=verbose)

    if verbose:
        print("P dimensions:", P.nrows(), "x", P.ncols())

    Psi_npz = load_Psi_npz_checked(psi_file, expected_n=P.nrows(), verbose=verbose)

    if verbose:
        print("converting Psi to Sage sparse integer matrix ...")
    Psi = scipy_csr_to_sage_sparse_int(Psi_npz)

    if verbose:
        print("expanding P over a QQ-basis of Q(zeta_36) ...")
    Ptilde = expand_K_matrix_to_QQ(P)

    if verbose:
        print("Ptilde dimensions:", Ptilde.nrows(), "x", Ptilde.ncols())

    if check_basis_expansion:
        check_expansion(P, Ptilde, K, z, num_rows=min(3, P.nrows()), num_cols=min(3, P.ncols()))

    if verbose:
        print("solving Psi * X = Ptilde over QQ ...")
    PsiQ = Psi.change_ring(QQ)
    Xtilde = solve_right_nonsingular_batched(PsiQ, Ptilde, batch_cols=rhs_batch_cols, verbose=verbose)

    Braw = Xtilde.transpose()

    if verbose:
        print("Braw dimensions:", Braw.nrows(), "x", Braw.ncols())
        print("computing QQ-row-space basis of Braw ...")

    Bq = Braw.row_space(QQ).basis_matrix()

    if verbose:
        print("QQ row-space rank =", Bq.nrows())
        print("clearing denominators and p-saturating ...")

    if saturation_primes is None:
        saturation_primes = [2, 3]

    B_aff, D = clear_denominators_and_psaturate_rows(
        Bq,
        primes=saturation_primes,
        proof=proof,
        verbose=verbose
    )

    if verbose:
        print("denominator multiplier D =", D)
        print("B_aff dimensions after p-saturation:", B_aff.nrows(), "x", B_aff.ncols())

    bad = audit_small_primes(B_aff, bound=29)
    if verbose:
        print("small primes where mod-p rank still drops:", bad)

    if verbose:
        print("computing affine Gram matrix G_aff = B_aff * Psi * B_aff^T ...")
    G_aff = gram_from_basis_and_sparse_psi(B_aff, Psi_npz)

    if verbose:
        print("G_aff dimensions:", G_aff.nrows(), "x", G_aff.ncols())

    # quotient by radical / cycles at infinity inside this lattice
    if verbose:
        print("quotienting by the saturated integral radical (cycles at infinity inside the lattice) ...")
    R_coeff, B_rad, B_proj, G_proj, U, G_new = quotient_by_radical_via_unimodular_change(
        B_aff, G_aff, proof=proof, verbose=verbose
    )

    if verbose:
        print("B_rad dimensions:", B_rad.nrows(), "x", B_rad.ncols())
        print("B_proj dimensions:", B_proj.nrows(), "x", B_proj.ncols())
        print("G_proj dimensions:", G_proj.nrows(), "x", G_proj.ncols())

    if expected_projective_rank is not None and B_proj.nrows() != expected_projective_rank:
        raise RuntimeError(
            "Expected a projective quotient rank-{} lattice, but got rank {}."
            .format(expected_projective_rank, B_proj.nrows())
        )

    if save_txt:
        with open(B_aff_output, "w") as f:
            f.write(str(B_aff.rows()))
        with open(G_aff_output, "w") as f:
            f.write(str(G_aff.rows()))
        with open(B_rad_output, "w") as f:
            f.write(str(B_rad.rows()))
        with open(B_proj_output, "w") as f:
            f.write(str(B_proj.rows()))
        with open(G_proj_output, "w") as f:
            f.write(str(G_proj.rows()))

        if verbose:
            print("wrote affine and projective quotient outputs to txt")

    if save_sobj:
        G_aff.save(G_aff_sobj)
        G_proj.save(G_proj_sobj)
        if verbose:
            print("saved G_aff and G_proj to sobj")

    return {
        "B_aff": B_aff,
        "G_aff": G_aff,
        "B_rad": B_rad,
        "B_proj": B_proj,
        "G_proj": G_proj,
        "radical_rank": B_rad.nrows(),
        "affine_rank": B_aff.nrows(),
        "projective_rank": B_proj.nrows(),
    }


# ------------------------------------------------------------
# 10) Example call
# ------------------------------------------------------------

out = compute_lattice_from_h22_periods_Qzeta36_with_infinity_quotient(
    period_file="matrixdegree36homologyh22even.txt",
    psi_file="Psi.npz",
    expected_affine_rank=18,
    expected_projective_rank=None,
    saturation_primes=[2, 3],
    rhs_batch_cols=4,
    verbose=True,
    save_txt=True,
    save_sobj=True,
    B_aff_output="B_affine.txt",
    G_aff_output="G_affine.txt",
    G_aff_sobj="G_affine.sobj",
    B_proj_output="B_projective.txt",
    G_proj_output="G_projective.txt",
    G_proj_sobj="G_projective.sobj",
    B_rad_output="B_radical.txt",
    proof=False,
    check_basis_expansion=True,
)

print("done")
print("affine rank:", out["affine_rank"])
print("radical rank:", out["radical_rank"])
print("projective quotient rank:", out["projective_rank"])
print("G_aff size:", out["G_aff"].nrows(), "x", out["G_aff"].ncols())
print("G_proj size:", out["G_proj"].nrows(), "x", out["G_proj"].ncols())
```
