```python
#Here we compute the gammas appearing in the scalar product
def compute_Gamma(h22):
    '''
    Computes the product of gammas functions appearing in the scalar product
    result: result, stores the product of gammas functions
    prod: product of gamma functions, symbolic variable
    '''
    result=[]
    prod=1
    for i in range(len(h22)):
        for j in range(len(h22[i])):
            if j!=(len(h22[i])-1) and j!=(len(h22[i])-2) and j!=(len(h22[i])-3):
                prod=prod*gamma((h22[i][j]+1)/36)
            if j==(len(h22[i])-3):
                prod=prod*gamma((h22[i][j]+1)/4)  
            if j==(len(h22[i])-2):
                prod=prod*gamma((h22[i][j]+1)/3)
            if j==(len(h22[i])-1):
                prod=prod*gamma((h22[i][j]+1)/3)
        result.append(prod.full_simplify())
        prod=1
    return result
fullgammas=compute_Gamma(h22even)
print(fullgammas) #only a single product of gammas in a pair of complex conjugates
#note that to scale up, we can simply use euler's reflection formula to proceed in linear time
#this is good enough for now
```

```python
#the function that does not depend on beta' and can therefore be introduced in the scalar product
def compute_xi(h22):
    '''
    Compute the xi term
    result: stores the results
    prod: stores the product of the xi terms.
    '''
    result=[]
    prod=1
    for i in range(len(h22)):
        for j in range(len(h22[i])):
            if j!=(len(h22[i])-1) and j!=(len(h22[i])-2) and j!=(len(h22[i])-3):
                prod=prod*(e^(2*I*pi/36*(h22[i][j]+1))-1)
            if j==(len(h22[i])-3):
                prod=prod*(e^(2*I*pi/4*(h22[i][j]+1))-1)
            if j==(len(h22[i])-2):
                prod=prod*(e^(2*I*pi/3*(h22[i][j]+1))-1)
            if j==(len(h22[i])-1):
                prod=prod*(e^(2*I*pi/3*(h22[i][j]+1))-1)
        result.append(prod.full_simplify())
        prod=1
    return result
fullxi=compute_xi(h22even)
print(fullxi)
#only a single xi for every pair of complex conjugate
#similarly with the gammas, there are not that many xi to compute in general so this is good enough
#although not optimal ( permutations induce some redundancy )
```

```python
def PeriodMatrix(L, beta_p, z, mults):
    """
    mults[l] = d // m_l  (embedding factor so that ζ_{m_l} = z**mults[l])

    Returns the matrix with rows i over L[i] and columns k over beta_p[k].

    Implements:
      ∏_{l} ( ζ_{m_l}^{(β_l+1)(β'_l+1)} − ζ_{m_l}^{β_l(β'_l+1)} )
    where β_l = beta_p[k][l] and β'_l = L[i][l].
    """
    Zvec = []
    for k in range(len(beta_p)):        # columns over beta_p[k]
        col = []
        for i in range(len(L)):         # rows over L[i]
            y = 1
            for l in range(len(L[i])):  # product over l
                c  = mults[l]           # = d/m_l
                bp  = beta_p[k][l]       # β_l
                b = L[i][l]            # β'_l
                # (z^(c*(b+1)*(bp+1)) - z^(c*b*(bp+1)))
                y *= (z ** (c * (bp) * (b + 1))) #- z ** (c * b * (bp + 1))) if not factorized
            col.append(y)
        Zvec.append(col)
    return Zvec
```

```python
d = K.degree() 
def coeffs_power_QQ(x):
    """
    Return the QQ^d power-basis coeff vector of x (w.r.t. the generator of its parent field).
    """
    # Coerce to K so .list() uses the power basis of z
    xK = K(x)
    L = list(xK.list())
    L += [0] * (d - len(L))
    return vector(QQ, L)
def coeffs_power_ZZ(x):
    v = coeffs_power_QQ(x)
    if any(c.denominator() != 1 for c in v):
        raise TypeError(f"Non-integral power-basis coefficients for {x}: {v}")
    return vector(ZZ, map(ZZ, v))

def build_Atilde(A):
    m, n = A.nrows(), A.ncols()
    blocks = []
    for j in range(n):
        B = matrix(ZZ, [list(coeffs_power_ZZ(A[i, j])) for i in range(m)])  # m x d
        blocks.append(B)
    return block_matrix(1, n, blocks)  # m x (d*n)

def split_into_blocks_of_4(Atilde_basis):
    """
    Split Atilde_basis into a list of m x 4 blocks [C_0, C_1, ..., C_{n-1}].
    """
    m, N = Atilde_basis.nrows(), Atilde_basis.ncols()
    n = N // d
    return [Atilde_basis[:, d*j:d*(j+1)] for j in range(n)]


txt = open("matrixdegree36homology.txt").read()
A = Matrix(K, sage_eval(txt, locals={'z': z}))
Atilde = build_Atilde(A)
```

```python
K = CyclotomicField(72)     # Field Q(zeta_12)
z = K.gen()

def row144_to_K12(r):
    """
    r: a Sage vector or list of length 144 (entries in QQ/ZZ)
    returns: vector in K^12 where each coordinate is sum_{k=0..11} r[12*j+k] * z^k
    """
    if len(r) != 144:
        raise ValueError("Expected a row of length 144")
    return vector(K, [
        sum(K(r[12*j + k]) * (z**(2*k)) for k in range(12))
        for j in range(len(r)/12)
    ])

def mat_to_K(M):
    """
    M: a matrix over QQ (or ZZ) with 144 columns.
    returns: a matrix over K with 12 columns (apply row144_to_K12 to each row).
    """
    if M.ncols() != 144:
        raise ValueError("Expected a matrix with 144 columns")
    return Matrix(K, [row144_to_K12(list(r)) for r in M.rows()])
```

```python
# ---------- Setup ----------
K.<z> = CyclotomicField(72)

def bar(x):            # complex conjugation in K (sends z -> z^-1)
    return x.conjugate()

def inner_scalar(a,b): # ⟨a,b⟩ on K
    return ((a*bar(b) + bar(a)*b)).trace() / 24

def inner_rows(u, v):
    return sum(inner_scalar(u[0,k], v[0,k]) for k in range(u.ncols()))

def gram_from_rows(R): # Gram matrix of the given rows
    m, n = R.nrows(), R.ncols()
    G = Matrix(QQ, m, m)
    for i in range(m):
        for j in range(m):
            s = 0
            for k in range(n):
                a = R[i,k]; b = R[j,k]
                s += (a*bar(b) + bar(a)*b).trace()
            G[i,j] = QQ(s) / 24
    return G

def dual_rows(R):
    """
    Input:  R — matrix over K whose rows are your vectors
    Output: D — matrix over K whose rows form the dual basis:
             <R[i], D[j]> = delta_ij
    """
    G = gram_from_rows(R)          # Gram over QQ
    Ginv = G.inverse()             # exact inverse over QQ
    return (Ginv.change_ring(K)) * R  # linear combo of the original rows


# ---------- COMPUTE THE DUAL ----------
D = dual_rows(CK)

# Verify orthonormality (should print the identity matrix):
m = CK.nrows()
Chk = Matrix(QQ, m, m, lambda i,j: inner_rows(CK[i,:], D[j,:]))
print("Check <CK[i], D[j]> = δ_ij ?")
print(Chk)
```

```python

# Cyclotomic field and basic generators
K.<z> = CyclotomicField(72)

# Algebraic replacements inside Q(zeta_72)
ii   = z^18                 # i
rt2  = z^9 + z^(-9)         # sqrt(2)
rt3  = z^6 + z^(-6)         # sqrt(3)

# Exponentials and sine in terms of zeta_72
e1 = z^22                   # exp( (11/18)*i*pi )
e2 = z^8                    # exp( (2/9)*i*pi )
s  = (z^25 - z^(-25)) / (2*ii)   # sin(25*pi/36)

# The two algebraic linear factors from the denominator
A = ((9*ii + 9)*rt3 + 9*ii - 9)*(e1 + e2) - (3*ii + 3)*rt3 + 3*ii + 9
B = ((9*ii - 9)*rt3 + 9*ii + 9)*(e1 + e2) + 6*rt3 - 6

# After Euler reflection:
# Gamma(25/36)^3 * Gamma(11/36)^3 = (pi / sin(25*pi/36))^3
# Gamma(2/3)^2 * Gamma(1/3)^2 = (2*pi/sqrt(3))^2
# Gamma(3/4) * Gamma(1/4) = pi*sqrt(2)
# This cancels pi^6 and yields the constant  (3/4) / sqrt(2) times sin^3.
const = Integer(161243136)  # = 214990848 * 3 / 4

# Final element in Q(zeta_72)
E = (const * (rt3 - ii) * s^3) / (rt2 * A * B)
E = K(E)                  # ensure as element of the cyclotomic field

# Show the result
E.trace()/24
```

```python
from itertools import product
C=2*E
def alpha_from_coeffs(a):
    # a = (a0,...,a_{d-1}) in Z^d
    if len(a) != len(w1):
        raise ValueError("len(a) must equal len(w1)")
    return sum(ai*wi for ai, wi in zip(a, w1))

def beta_from_coeffs(b):
    # b = (b0,...,b_{d-1}) in Z^d
    if len(b) != len(w4):
        raise ValueError("len(b) must equal len(w4)")
    return sum(bi*wi for bi, wi in zip(b, w4))

def plain_val(a, b, scale=True):
    v = alpha_from_coeffs(a) * beta_from_coeffs(b)
    return (C*v) if scale else v

def is_integer_in_Q(x):
    # x lies in K; test if it's a rational integer
    return (x in QQ) and QQ(x).denominator() == 1

def min_trace_over_box(n, *, diagonal=True, scale=True):
    """
    Search a,b with coords in [-n,n]^{d}.
    If diagonal=True, uses b=a (far cheaper); else enumerates all pairs (a,b).
    Returns the minimum of (1/4)*trace(plain_val(a,b,scale)).
    """
    d1, d2 = len(w1), len(w4)
    if d1 != d2:
        raise ValueError("w1 and w4 must have the same length")
    R = range(-n, n+1)

    t_min = None
    total = 0

    if diagonal:
        for a in product(R, repeat=d1):
            v = plain_val(a, a, scale=scale)
            if v != 0:
                t = (v.trace())/24
                t_min = t if t_min is None or t < t_min else t_min
            total += 1
    else:
        for a in product(R, repeat=d1):
            for b in product(R, repeat=d2):
                v = plain_val(a, b, scale=scale)
                if v != 0:
                    t = (v.trace())/4
                    t_min = t if t_min is None or t < t_min else t_min
                total += 1

    print("Tuples checked:", total)
    return t_min

# === Example usage ===
 
n = 5
tmin = min_trace_over_box(n, diagonal=True, scale=True)
print("min (1/4)*trace =", tmin)
```
